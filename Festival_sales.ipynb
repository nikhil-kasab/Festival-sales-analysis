# Exploratory Data Analysis

## GENDER

```python
sns.barplot(x= 'Gender' , data = df)
```

```python
sns.set(rc={'figure.figsize':(15,5)})
```

```python
ax = sns.countplot(x = 'Gender',hue = 'Gender' , data = df)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)


sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen,hue = 'Gender')
```

*from the above graph we can clearly see that there are more of female buyers than that of men*

## AGE

```python
# genderwise age group
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)     # for representing the count  on bars
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount' , ascending=False)

sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age , hue = 'Age Group')
```

*most number of purchases are done by females of age group 26-35 and least are done by children of age group 0-17*

## STATE

```python
# total number of orders from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)


sns.set(rc={'figure.figsize':(20,7)})
sns.barplot(data = sales_state, x = 'State',y= 'Orders' , hue = 'State')
```

```python
# total amount/sales from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,7)})
sns.barplot(data = sales_state, x = 'State',y= 'Amount' ,hue = 'State')
```

*from above graphs we can clearly see most orders are from UP followed by Maharashtra and Karnataka respectively*

# product catagory

```python
ax = sns.countplot(data = df , x = 'Product_Category' , hue = 'Product_Category' )

sns.set(rc={'figure.figsize':(25,9)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
category = df.groupby(['Product_Category'],as_index = False) ['Amount'].sum().sort_values(by = 'Amount',ascending=False)

sns.set(rc={'figure.figsize':(35,10)})


sns.barplot(data = category , x='Product_Category' , y = 'Amount' , hue = 'Product_Category')
```

*although the most products sold are clothings but most amount is being spent on food followed by clothings and electronics respectively*

```python
# product_ids of top 10 most selling products

top_sellers = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.barplot(data = top_sellers , x = 'Product_ID',y= 'Orders' , hue = 'Product_ID')
```
